# Segmentación de las estaciones

```python
from sklearn.cluster import KMeans

data = stations[['lat', 'lon']]

sse = {}
for k in range(1, 30):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(data)
    data["clusters"] = kmeans.labels_
    #print(data["clusters"])
    sse[k] = kmeans.inertia_ # Nota: Suma de distancias de las muestras hacia el centroide más cercano
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```

```python
data = stations[['lat', 'lon']]

kmeans = KMeans(n_clusters=5, max_iter=1000).fit(data)
 
data["clusters"] = kmeans.labels_

#Getting the Centroids
centroids = kmeans.cluster_centers_
u_labels = np.unique(data["clusters"])

centroids_x = centroids[:,0]
centroids_y = centroids[:,1]
 
plt.scatter(centroids_x,centroids_y,marker = "x", s=150,linewidths = 5, zorder = 10, c=['green', 'gray','blue','yellow','orange'])
```

## La técnica utilizada para la segmentación fue el algoritmo de K-means o K-medias, el cual se utiliza para clustering de manera no supervisada. La desventaja de este algoritmo es que se debe definir un hiperparámetro k, el cual es para definir el número de clústers o segmentaciones y las segmentaciones tendrían que ser de manera arbitrarias, pero gracias a un método denominado el "método del codo", el cual consiste en correr el algoritmo k-means para diferentes valores de k, cuando la curva ya no descienda más y empiece a tener una tendencia hacia cero cuando el límite tiende a infinito positivo, ese punto es considerado el valor óptimo de k clusters o segmentaciones. Para este caso, como puede verse en la figura previa, el valor óptimo de k es 5. Las variables utilizadas por el algoritmo fue la latitud y longitud de las estaciones, debido a la alta correlación entre las variables de estación de retiro y arribo mostrado en la respuesta anterior, y dicha correlación tiende a dar con las distancias entre ellas.
